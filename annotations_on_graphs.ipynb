```
import numpy as np
import matplotlib.pyplot as plt
```

```
x = np.arange(-5,6,)
y = x**2

fig,ax = plt.subplots()

ax.plot(x,y,'ko-',markerfacecolor='b',markersize=10)
ax.annotate('collar',(x[2],y[2]),fontsize=15,arrowprops=dict(),
            xycoords='data',xytext=(0,10),horizontalalignment='center')

plt.show()
```

```
y = x**2 + np.cos(x)*10

minpnt = np.argmin(y)
txt = 'minimun: (%g,%.2f)'%(x[minpnt],y[minpnt])

fig,ax = plt.subplots()
ax.plot(x,y,'ko-',markerfacecolor='purple',markersize=15)
ax.annotate(txt,(x[minpnt],y[minpnt]),fontsize=15,
            xycoords='data',xytext=(x[int(len(x)/2)],np.max(y)*.8),horizontalalignment='center',
            arrowprops=dict(color='green',arrowstyle='wedge'))

ax.set_xlabel('x')
ax.set_ylabel('f(x) = x^2 + 10cos(x)')
plt.show()
```
