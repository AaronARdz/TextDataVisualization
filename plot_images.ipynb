```
import matplotlib.pyplot as plt
import numpy as np

m = 6
n = 9

matrix = np.random.randint(10,size=(m,n))
print(matrix)

plt.imshow(matrix)

for i in range(m):
  for j in range(n):
    plt.text(j,i,str(matrix[i,j]),horizontalalignment='center',fontsize=10)


plt.colorbar()
plt.show()
```

```
import seaborn as sns
# Exercise: 
# 1) create hilbert matrix(10x10) H[i,j] = 1 / (i+j-1)
# 2) visualize the hilbert matrix using seaborn

x = 10

H = np.zeros((x,x))

for i in range(1,x+1):
  for j in range(1,x+1):
    H[i-1,j-1] = 1/(i+j-1)

# 2) visualize the Hilber matrix using seaborn

sns.heatmap(H, vmin=0, vmax=.3)
plt.show()
```
